```python
import os
import numpy as np
import matplotlib.pyplot as plt

cells_Name = [
    "bNAC219_L1_NGCDA_095997ce92",
    "bNAC219_L1_NGCDA_e7cec642c3",
    "bNAC219_L1_NGCDA_46b45974f4",
    #"bNAC219_L1_NGCDA_b030a58a66",
    #"bNAC219_L1_NGCDA_3d9c976fde",
]

base_dir = "/home/lascon/Documents/241165/AberraEtAl2018/cells"

plt.figure(figsize=(6, 4))

for cell_num, cell_Name in enumerate(cells_Name, start=1):
    cell_path = os.path.join(base_dir, f"L1_NGC-DA_bNAC219_{cell_num}")
    rates_file = os.path.join(cell_path, "rates.dat")

    # rates.dat format: idx  amp_nA  rate_Hz
    rates_data = np.loadtxt(rates_file, comments="#")
    amps = rates_data[:, 1]
    rates_hz = rates_data[:, 2]

    plt.plot(amps, rates_hz, "o-", lw=2, label=cell_Name)

plt.xlabel("Injected current (nA)")
plt.ylabel("Firing rate (Hz)")
plt.title("FI curve  L1 NGC-DA bNAC219")
plt.grid(True)
plt.legend(fontsize=7)
plt.tight_layout()
plt.show()
```
